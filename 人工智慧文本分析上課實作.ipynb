```python
import matplotlib.pyplot as plt #繪圖函式庫
import numpy as np
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'gx') #綠色，x的點
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'r--') #紅色，--虛線

#取亂數
X=1+np.arange(30)/10
delta=np.random.uniform(low=-0.1,high=0.1,size=(30,))
#帶入公式
Y=0.3*X-0.3+delta
plt.plot(X,Y,'bo') #藍色的圓點
plt.xlabel('X') #設定顯示X文字
plt.ylabel('Y') #設定顯示Y文字
plt.show() #繪製圖表
```

```python
import matplotlib.pyplot as plt #繪圖函式庫
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'gx') #在x=1,y=0四個位置繪製(g)色標記
plt.plot([1,2,3,4],[0,0.3,0.6,0.9],'r--') #繪製(r)紅色-標記也就是線(趨勢線)
plt.axis([0,5,0,1]) #圖表大小範圍，寬度由0到5，高度由0到1
plt.ylabel('Y') #設定顯示Y文字
plt.xlabel('X') #設定顯示X文字
plt.legend(('price','passenger'),loc='upper right')
plt.show() #繪製圖表
```

```python
import pandas as pd
from sklearn import linear_model #繪圖函式庫
import matplotlib.pyplot as plt #繪圖函式庫

#read data
dataframe=pd.read_fwf('brain_body.txt') #讀取txt的數據
x_values=dataframe[['Body']] #身體的數據
y_values=dataframe[['Brain']] #大腦的數據

#訓練
body_reg=linear_model.LinearRegression() #使用線性迴歸類型

body_reg.fit(x_values,y_values)
pre=body_reg.predict(x_values)

#訓練預測
print(body_reg.predict(pd.DataFrame(data=[170])))

#圖形化
plt.scatter(x_values,y_values)
plt.plot(x_values,pre)
plt.show()
```

```python
import jieba
text='我來到台灣新北市的淡江大學'
```

```python
print('預設:','|'.join(jieba.cut(text,cut_all=False,HMM=True)))
print('全關閉:','|'.join(jieba.cut(text,cut_all=False,HMM=False)))
print('全關閉:','|'.join(jieba.cut(text,cut_all=True,HMM=True)))
```

```python
text='富邦金證實董事長蔡明興確診 由蔡明忠代理職務'
print('預設:','|'.join(jieba.cut(text,cut_all=False,HMM=True)))
```

```python
jieba.add_word("富邦金")
jieba.add_word("董事長")
print('預設:','|'.join(jieba.cut(text,cut_all=False,HMM=True)))
```

```python
import jieba.posseg as pseg
words=pseg.cut(text)
for word,flag in words:
   print(word,flag)
```

```python
import jieba.analyse
tags=jieba.analyse.extract_tags(text,topK=6)
print(tags)
```
